```python
#init a 16-values vector, this is our input
input_state = np.zeros(16)
input_state[0] = 1 #so we have v=(1,0,0,...,0): |0000>

#step 1: we perform H2 on the first 2 qubits as the last 2 remain |00> for simon
transformation = np.kron(H2, I2)
state_after_step_1 = transformation @ input_state # HxHxIxI|0000>

#step 2: mlt with Uf
state_after_step_2 = U_f @ state_after_step_1 # Uf|state_after_step_1>
```

# Task 1

A: Given the measurements, the range is {10,11}

```python
## prob of measuring 00
operator = np.kron(I2, p_00)
temp = operator @ state_after_step_2
prob_00 = np.linalg.norm(temp)**2
print(f"Probability of 00: {prob_00}")

## prob of measuring 01
operator = np.kron(I2, p_01)
temp = operator @ state_after_step_2
prob_01 = np.linalg.norm(temp)**2
print(f"Probability of 01: {prob_01}")

## prob of measuring 10
operator = np.kron(I2, p_10)
temp = operator @ state_after_step_2
prob_10 = np.linalg.norm(temp)**2
print(f"Probability of 10: {prob_10}")

## prob of measuring 11
operator = np.kron(I2, p_11)
temp = operator @ state_after_step_2
prob_11 = np.linalg.norm(temp)**2
print(f"Probability of 11: {prob_11}")
```
